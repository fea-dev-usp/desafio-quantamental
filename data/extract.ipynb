```python
import re
tickers_clean = re.sub('\s+', '', tickers)
tickers_split = tickers_clean.split(',')
tickers = list(set(tickers_split))
```

```python
start = datetime(2009, 11, 1)
end = datetime(2019, 12, 31)
```

```python

dfs = []
for i, ticker in enumerate(tickers):
    try:
        df_aux = web.DataReader(ticker, 'av-monthly-adjusted',start=start, end=end, api_key='EJHD53RD3NC1QNCV')

        df_aux = pd.DataFrame({ticker:df_aux['adjusted close']})

        df_aux.index = pd.to_datetime(df_aux.index, format='%Y-%m-%d')

        df_aux.index = df_aux.index.to_period('M')
        dfs.append(df_aux)

        if i % 5 == 0:
            time.sleep(61)
    except:
        print('Ticker falho: ', ticker)
```

```python
df = pd.concat(dfs, axis = 1)
```

```python
for column in df.columns:
    print(column, df[column].isna().sum())
```
